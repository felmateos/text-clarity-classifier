### DataFrame

```python
import pandas as pd
import numpy as np
#df = pd.read_csv('./datasets/DfPos.csv')
df = pd.read_csv('./datasets/DfSLemma.csv')
```

### Vectorize

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(df['textos'])

X_train_Tfidf = Tfidf_vect.transform(df['textos'])
X_test_Tfidf = Tfidf_vect.transform(df['textos'])
```

```python
y = df['Class']
```

### Model

```python
from sklearn import svm
```

```python

SVM = svm.SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(SVM, X_train_Tfidf, y, cv=10)

print(scores)
scores.mean()
```
